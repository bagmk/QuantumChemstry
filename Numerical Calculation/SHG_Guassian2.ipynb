```python
import numpy as np
import math
import scipy.linalg as la
import matplotlib
import matplotlib.pyplot as plt
```

# SHO Potential

```python
N=100
n = list(range(0, N))
s = np.zeros(N)
h = np.zeros((N,N))
o = np.zeros((N,N))
for i in n:
    s[i] = -25.0+ (i)*0.5

for i in n:
    for j in n:
        ss = (s[i]-s[j])**2
        #print(ss)
        o[i][j] = math.exp(-0.5*ss)
        t = math.exp(-0.5*ss)*0.5*(1-ss)
        p = 0.5* math.exp(-0.5*ss)*0.25*(1+(s[i]+s[j])**2)
        h[i][j] = t + p
#print(h) 
#eigvals, eigvecs = np.linalg.eigh(h)
eigvals, eigvecs = la.eigh(h,o)

print(eigvals)


```

# Gaussian Potential

```python
N=100
n = list(range(0, N))
s = np.zeros(N)
h = np.zeros((N,N))
o = np.zeros((N,N))
urang=[10**((i-15)/15) for i in range(31)]

gsp=[]
gsp1=[]
gsp2=[]
gsp3=[]
gsp4=[]
gsp5=[]
gsp6=[]
gsp7=[]
for u in urang:
    for i in n:
        s[i] = -25.0+ (i)*0.5

    for i in n:
        for j in n:
            ss = (s[i]-s[j])**2
            #print(ss)
            o[i][j] = math.exp(-0.5*ss)
            t = math.exp(-0.5*ss)*0.5*(1-ss)
            p = (2/(2+u)**0.5) *math.exp(-0.5*ss)*math.exp(-((s[i]-s[j])**2+u*s[i]**2+u*s[j]**2)/(2+u))
            h[i][j] = t + p
    #print(h) 
    #eigvals, eigvecs = np.linalg.eigh(h)
    eigvals, eigvecs = la.eigh(h,o)
    gsp.append(eigvecs[0])
    gsp1.append(eigvals[1])
    gsp2.append(eigvals[2])
    gsp3.append(eigvals[3])
    gsp4.append(eigvals[4])
    gsp5.append(eigvals[5])
    gsp6.append(eigvals[6])
    gsp7.append(eigvals[7])
```
